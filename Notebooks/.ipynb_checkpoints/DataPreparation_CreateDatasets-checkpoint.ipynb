```python
## Import game list and shot list
path_games = '/work/users/ca439qyki/R/Kinexon/Data/Goals/'
game_list_df = pd.read_csv('{path}{filename}'.format(path=path_games, filename='game_list.csv'))
shot_list_df = pd.read_csv('{path}{filename}'.format(path=path_games, filename='shot_list.csv'))

## only get matches with available positional data
# teams = ['SC DHFK Leipzig', 'MT Melsungen', 'Bergischer HC']
teams = ['SC DHFK Leipzig','Bergischer HC']
game_list_df = game_list_df[game_list_df['home_team'].isin(teams) & game_list_df['away_team'].isin(teams)]

## drop unused columns and not filled rows
game_list_df = game_list_df.drop(game_list_df.columns[0], axis=1)
# shot_list_df = shot_list_df.drop(shot_list_df.columns[[0,13,15,16,17,18,19,21]], axis=1)
# shot_list_df = shot_list_df.drop(shot_list_df.columns[[0,13,15,16]], axis=1)
shot_list_df = shot_list_df[((shot_list_df['game_clock']!='30:00') &(shot_list_df['validated'] != 0))] ## drop all shots from the halftime
shot_list_df = shot_list_df[shot_list_df['game_clock'].notna()] # drop all shot pre game

## add columns to connect shots to kinexon data before splitting it up per game
shot_list_df['timestamp_dt'] = pd.to_datetime(shot_list_df['timestamp'])
shot_list_df['merge_time_shots'] = shot_list_df['timestamp_dt'].apply(lambda x: int(x.value / 1000000))
## Games in Season 2019-2020 does not have a match_id => fill this match id with the match id from the gamelist
shot_list_df.loc[(shot_list_df['match_id'].isna() & shot_list_df['team'].isin(teams) & shot_list_df['timestamp_dt'].dt.date.isin(pd.to_datetime(game_list_df['date']).dt.date)), 'match_id'] = 18949282
shot_list_df = shot_list_df[shot_list_df['team'].isin(teams) & shot_list_df['match_id'].isin(game_list_df['match_id'])]

## add shots per game to the shot row in the dataframe
game_list_df['shots'] = game_list_df['match_id'].apply(lambda x: shot_list_df[shot_list_df['match_id'] == x])
```

```python
## define utils functions
def calc_game_score(home, success, score_home= 0, score_away= 0):
    if success:
        if home:
            score_home += 1
        else:
            score_away += 1
    return score_home, score_away

def standarderize_coordiates(home, halftime, x, y, saison):
    if saison != '2020_2021':
        if not halftime:
            if home:
                return x, y
            else: 
                return x*(-1), y*(-1)
        else:
            if home:
                return x*(-1), y*(-1)
            else:
                return x, y
    else:
        if not halftime:
            if home:
                return x*(-1), y*(-1)
            else: 
                return x, y
        else:
            if home:
                return x, y
            else:
                return x*(-1), y*(-1)

def calculate_angle_to_mid(x, y):
    S = np.array([x, y])
    Mid = np.array([-20, 0])
    MidS = np.array([x, 0])
    d_v = math.dist(S, Mid) 
    d_w = math.dist(MidS, Mid)
    v = S - Mid
    w = MidS - Mid
    return np.arccos(np.dot(v,w)/(d_v * d_w))

def calculate_angle_from_shooter_to_gk(x_Shot, y_Shot, x_gk, y_gk):
    S = np.array([x_Shot, y_Shot])
    if x_gk < 0:
        Mid = np.array([-20, 0])
    else:
        Mid = np.array([20, 0])
    MidS = np.array([x_gk, y_gk])
    d_v = math.dist(S, Mid) 
    d_w = math.dist(MidS, Mid)
    v = S - Mid
    w = MidS - Mid
    return np.arccos(np.dot(v,w)/(d_v * d_w))

def calculate_visible_angle(x,y):
    S = np.array([x, y])
    G1 = np.array([-20,-1.5])
    G2 = np.array([-20, 1.5])
    d_v = math.dist(S, G1)
    d_w = math.dist(S, G2)
    v = G1 - S
    w = G2 - S
    return np.arccos(np.dot(v,w)/(d_v * d_w))

def calculate_distance(x1,y1, x2, y2):
    P1 = np.array([x1, y1])
    P2 = np.array([x2, y2])
    return math.dist(P1, P2)

def round_to_next_n(x, n=50):
    return int(n * round(float(x)/n))

def players_in_shot_angle(x_shot, y_shot, home, halftime, pos_x_list, pos_y_list, saison):
    if saison != '2020_2021':
        if ((home) and (not halftime)) or ((not home) and (halftime)):
            polygon = Polygon([(x_shot, y_shot), (-20, -1.50), (-20, 1.50)])
        else:
            polygon = Polygon([(x_shot, y_shot), (20, -1.50), (20, 1.50)])
    else:
        if ((home) and (not halftime)) or ((not home) and (halftime)):
            polygon = Polygon([(x_shot, y_shot), (20, -1.50), (20, 1.50)])
        else:
            polygon = Polygon([(x_shot, y_shot), (-20, -1.50), (-20, 1.50)])
    
    i = 0
    for pos_x, pos_y in zip(pos_x_list, pos_y_list):
        point = Point(pos_x, pos_y)
        i += polygon.contains(point)    
        
    return i

func_round = np.vectorize(round_to_next_n)
func_coord = np.vectorize(standarderize_coordiates)
func_visible_angle = np.vectorize(calculate_visible_angle)
func_angle_to_mid = np.vectorize(calculate_angle_to_mid)
func_game_score = np.vectorize(calc_game_score)
func_angle_shooter_gk = np.vectorize(calculate_angle_from_shooter_to_gk)
func_dist = np.vectorize(calculate_distance)
```

```python
##load mapping table of player ids
player_ids = pd.read_csv('/home/sc.uni-leipzig.de/az091jwou/expected_goals_handball/Data/players_sensors_map.csv', usecols=[1,2,3,4,5])
goalkeepers = player_ids[player_ids['goalie'] == 1]

shot_list_df = pd.DataFrame()

for idx, row in game_list_df.iterrows():
    if row['saison'] == '2019_2020':
        teams_dict = {'SC DHFK Leipzig':('DHFK', 'SC_DHfK_Leipzig'),'Bergischer HC': ('BHC', 'Bergischer')}
    else:
        teams_dict = {'SC DHFK Leipzig':('DHFK', 'SC_DHfK_Leip'),'Bergischer HC': ('BHC', 'Bergischer_HC')}
    
    ## import relevant Kinexon Data for the Game
    home_team = teams_dict.get(row['home_team'])
    away_team = teams_dict.get(row['away_team'])
    home_inertial_df = pd.read_csv(kinexon_inertial_path.format(short=home_team[0], season=row['saison'],gameday_number=row['game_day'], team_name_home=home_team[1], team_name_away=away_team[1]), sep=';', usecols=[0,2,9,10,11])
    home_pos_df = pd.read_csv(kinexon_positions_path.format(short=home_team[0], season=row['saison'],gameday_number=row['game_day'], team_name_home=home_team[1], team_name_away=away_team[1]), sep=';',usecols=[0,1,2,3,4,5,7,8,9,10,11,16,17])
    away_inertial_df = pd.read_csv(kinexon_inertial_path.format(short=away_team[0], season=row['saison'],gameday_number=row['game_day'], team_name_home=home_team[1], team_name_away=away_team[1]), sep=';', usecols=[0,2,9,10,11])
    away_pos_df = pd.read_csv(kinexon_positions_path.format(short=away_team[0], season=row['saison'],gameday_number=row['game_day'], team_name_home=home_team[1], team_name_away=away_team[1]), sep=';', usecols=[0,1,2,3,4,5,7,8,9,10,11,16,17])
    
    ## merge inertial and pos Dataframes for home and away
    away_inertial_df['merge_time'] = func_round(away_inertial_df['ts in ms'])
    away_pos_df['merge_time'] = away_pos_df['ts in ms']
    home_inertial_df['merge_time'] = func_round(home_inertial_df['ts in ms'])
    home_pos_df['merge_time'] = home_pos_df['ts in ms']
    away_game_data = pd.merge(away_pos_df, away_inertial_df, how='left', on=['merge_time', 'sensor id'], suffixes=(None, '_inertial'))
    home_game_data = pd.merge(home_pos_df, home_inertial_df, how='left', on=['merge_time', 'sensor id'], suffixes=(None, '_inertial'))
    # game_data = pd.concat([away_game_data, home_game_data], ignore_index=True)
    
    ## drop Rows where no x and y coordinates are given in the kinexon datasets
    home_game_data = home_game_data.dropna(subset=['x in m', 'y in m'])
    away_game_data = away_game_data.dropna(subset=['x in m', 'y in m'])
    
    ## Create timestamp for Kinexon-Data in ms to compare it with the shots
    home_game_data['merge_time_shots'] = func_round(home_game_data['ts in ms'], n=1000)
    away_game_data['merge_time_shots'] = func_round(away_game_data['ts in ms'], n=1000)
    
    ## Calculate moving average of the speed of the players (not really sure if i need it)
    # away_game_data['moving mean speed'] = away_game_data.groupby('mapped id')['speed in m/s'].transform(lambda x: x.rolling(16, 1).mean())
    # home_game_data['moving mean speed'] = home_game_data.groupby('mapped id')['speed in m/s'].transform(lambda x: x.rolling(16, 1).mean())
    
    ## Initialize last_shot to get the number of passes and seconds of ball posession
    last_shot_timestamp = starting_times.get(row['match_id'])
    
    ## Create features for each shot
    ## reverse sorting of match 18949282 because the last shot is first in the list
    shot_list_game = pd.DataFrame(row['shots'])
    if (row['match_id'] == 18949282):
        shot_list_game = shot_list_game.iloc[::-1]

    shot_list_game = shot_list_game[shot_list_game['merge_time_shots'] >= starting_times.get(row['match_id'])]
    shot_list_game['halftime'] = pd.to_timedelta(['00:'+time for time in shot_list_game['game_clock']]) > datetime.timedelta(minutes=30)
    shot_list_game['home'] = shot_list_game['team'].isin([row['home_team']])
    shot_list_game['norm_x_shot'], shot_list_game['norm_y_shot'] = func_coord(shot_list_game['home'], shot_list_game['halftime'], shot_list_game['shot_position_x'], shot_list_game['shot_position_y'], row['saison'])
    shot_list_game['visible_angle'] = func_visible_angle(shot_list_game['norm_x_shot'], shot_list_game['norm_y_shot'])
    shot_list_game['angle_to_mid'] = func_angle_to_mid(shot_list_game['norm_x_shot'], shot_list_game['norm_y_shot'])
    shot_list_game['score_home'], shot_list_game['score_away'] = func_game_score(shot_list_game['home'], shot_list_game['success'])
    shot_list_game['score_home'] = shot_list_game.groupby(['score_home']).ngroup().cumsum()
    shot_list_game['score_away'] = shot_list_game.groupby(['score_away']).ngroup().cumsum()
    shot_list_game.sort_values(['timestamp_dt', 'home'], inplace=True)
    shot_list_game['time_between_shots_s'] = shot_list_game.groupby('home')['timestamp_dt'].diff().dt.total_seconds()
    shot_list_game.sort_index(inplace=True)
    
    ## Initialize passes DataFrame for the game
    passes = pd.concat([away_game_data[(away_game_data['ball possession (id of possessed ball)'].notna()) & (away_game_data['ts in ms'] >= starting_times.get(row['match_id']))], 
                        home_game_data[(home_game_data['ball possession (id of possessed ball)'].notna()) & (home_game_data['ts in ms'] >= starting_times.get(row['match_id']))]])
    passes = passes.loc[passes['mapped id'].shift() != passes['mapped id']].sort_values('ts in ms')
    
    ## Initialize list with shots that were captured manually => shots that does not contain positional data of the shot
    idx_manual_shots = []
    
    ## Create Dataframe for Shots of the particular game
    for idx, shot in shot_list_game.iterrows():
        ## Get Kinexon-Data for the timestamp of the shot
        shot_away_pos = away_game_data[away_game_data['merge_time_shots'] == shot['merge_time_shots']]
        shot_home_pos = home_game_data[home_game_data['merge_time_shots'] == shot['merge_time_shots']]
        player = player_ids.loc[player_ids['league_id'] == shot['league_id']]['player_id'].values
        
        # Check if player mapping from player id to mapped id is available (implemented to find errors i the data)
        if player.size == 0:
            print(shot,'\n')
            break
        elif player.size > 1:
            print(shot,'\n')
            print(player)
            break
        
        ## Get Shooter and Goalkeeper Data (position, speed...)
        if shot['home']:
            shooter = shot_home_pos[(shot_home_pos['mapped id'] == player.item())]
            goalkeeper = shot_away_pos.loc[shot_away_pos['mapped id'].isin(goalkeepers['player_id'])]
            
        else:
            shooter = shot_away_pos[(shot_away_pos['mapped id'] == player.item())]
            goalkeeper = shot_home_pos.loc[shot_home_pos['mapped id'].isin(goalkeepers['player_id'])]
            
        ## Some manual captured shots does not math the kinexon data => Shot will be skipped
        if shooter.empty:
            print('No shooter found', idx)
            continue

        ## Get snapshot timestamp to get a frozen frame of all player positions
        ## maybe add some criterias to define the shot timestamp => easy method is to get the last timestamp of the shot timestamp but it seems to be inaccurate for some shots 
        # if pd.DataFrame(shooter.loc[shooter['ball possession (id of possessed ball)'].notna()]).empty:
        snapshot_shot_timestamp = shooter['ts in ms'].iloc[-1]
        # else:
        #     snapshot_shot_timestamp = shooter.loc[shooter['ball possession (id of possessed ball)'].notna()]['ts in ms'].iloc[-1]
        
        ## Get player positions to the calculated timestamp
        snapshot_home = shot_home_pos[shot_home_pos['ts in ms'] == snapshot_shot_timestamp]
        snapshot_away = shot_away_pos[shot_away_pos['ts in ms'] == snapshot_shot_timestamp]
        snapshot_goalkeeper = goalkeeper[goalkeeper['ts in ms'] == snapshot_shot_timestamp]
        snapshot_shooter = shooter[shooter['ts in ms'] == snapshot_shot_timestamp]
        
        ## Sometimes if no goalkeeper was found, the searched timestamp is not available => look for the timestamp 50ms earlier
        if (snapshot_goalkeeper.empty) & (~goalkeeper.empty):
            snapshot_goalkeeper = goalkeeper.tail(1)
            
        if len(snapshot_goalkeeper) > 1:
            snapshot_goalkeeper = snapshot_goalkeeper.head(1)
            
        ## Add values to shots that were tracked manually => Add position of the shooter as value for the shot coordinates
        if np.isnan(shot['shot_position_x']) and np.isnan(shot['shot_position_y']):
            shot['shot_position_x'] = snapshot_shooter['x in m'].values[0]
            shot['shot_position_y'] = -snapshot_shooter['y in m'].values[0]
            shot['norm_x_shot'], shot['norm_y_shot'] = standarderize_coordiates(shot['home'], shot['halftime'], shot['shot_position_x'], shot['shot_position_y'], row['saison'])
            shot_list_game.at[idx, 'shot_position_x'] = snapshot_shooter['x in m'].values[0]
            shot_list_game.at[idx, 'shot_position_y'] = -snapshot_shooter['y in m'].values[0]
            shot_list_game.at[idx, 'norm_x_shot'], shot_list_game.at[idx, 'norm_y_shot'] = standarderize_coordiates(shot['home'], shot['halftime'], shot['shot_position_x'], shot['shot_position_y'], row['saison'])
            shot_list_game.at[idx, 'distance'] = calculate_distance(shot['norm_x_shot'], shot['norm_y_shot'], -20, 0)
            shot_list_game.at[idx, 'visible_angle'] = func_visible_angle(shot['norm_x_shot'], shot['norm_y_shot'])
            shot_list_game.at[idx, 'angle_to_mid'] = func_angle_to_mid(shot['norm_x_shot'], shot['norm_y_shot'])
            idx_manual_shots.append(idx)
            
        ## Calculate distance from shooter to defenders to get the number of defenders in a given radius
        ## Get Number of Players between shooter and goal
        if shot['home']:
            snapshot_away['dist_to_shooter'] = func_dist(snapshot_away['x in m'], snapshot_away['y in m'], snapshot_shooter['x in m'], snapshot_shooter['y in m'])
            shot_list_game.at[idx, 'pressur_index'] = len(snapshot_away[(snapshot_away['dist_to_shooter'] <= 1) & (snapshot_away['group name'] != 'Ball')])
            shot_list_game.at[idx, 'dist_closest_def'] = snapshot_away[(snapshot_away['group name'] != 'Ball')].sort_values('dist_to_shooter').iloc[0]['dist_to_shooter']
            ## Maybe get only field players between => now it includes the goalkeeper as well
            shot_list_game.at[idx, 'players_in_shot_angle_from_shot'] = players_in_shot_angle(shot['shot_position_x'], shot['shot_position_y'],
                                                                                shot['home'], shot['halftime'], list(snapshot_away[(snapshot_away['group name'] != 'Ball')]['x in m']), list(snapshot_away[(snapshot_away['group name'] != 'Ball')]['y in m']),row['saison'])
            shot_list_game.at[idx, 'players_in_shot_angle_from_player'] = players_in_shot_angle(snapshot_shooter['x in m'].values[0], snapshot_shooter['y in m'].values[0], 
                                                                                shot['home'], shot['halftime'], list(snapshot_away[(snapshot_away['group name'] != 'Ball')]['x in m']), list(snapshot_away[(snapshot_away['group name'] != 'Ball')]['y in m']), row['saison'])
            shot_list_game.at[idx, 'powerplay'] = len(snapshot_away[(snapshot_away['group name'] != 'Ball') & (~snapshot_away['mapped id'].isin(goalkeepers))]) < len(snapshot_home[(snapshot_home['group name'] != 'Ball')& (~snapshot_home['mapped id'].isin(goalkeepers))])
        else:
            snapshot_home['dist_to_shooter'] = func_dist(snapshot_home['x in m'], snapshot_home['y in m'], snapshot_shooter['x in m'], snapshot_shooter['y in m'])
            shot_list_game.at[idx, 'pressur_index'] = len(snapshot_home[(snapshot_home['dist_to_shooter'] <= 1) & (snapshot_home['group name'] != 'Ball')])
            shot_list_game.at[idx, 'dist_closest_def'] = snapshot_home[(snapshot_home['group name'] != 'Ball')].sort_values('dist_to_shooter').iloc[0]['dist_to_shooter']
            ## Maybe get only field players between => now it includes the goalkeeper as well
            shot_list_game.at[idx, 'players_in_shot_angle_from_shot'] = players_in_shot_angle(shot['shot_position_x'], shot['shot_position_y'], 
                                                                                shot['home'], shot['halftime'], list(snapshot_home[(snapshot_home['group name'] != 'Ball')]['x in m']), list(snapshot_home[(snapshot_home['group name'] != 'Ball')]['y in m']), row['saison'])
            shot_list_game.at[idx, 'players_in_shot_angle_from_player'] = players_in_shot_angle(snapshot_shooter['x in m'].values[0], snapshot_shooter['y in m'].values[0], 
                                                                                shot['home'], shot['halftime'], list(snapshot_home[(snapshot_home['group name'] != 'Ball')]['x in m']), list(snapshot_home[(snapshot_home['group name'] != 'Ball')]['y in m']), row['saison'])
            shot_list_game.at[idx, 'powerplay'] = len(snapshot_away[(snapshot_away['group name'] != 'Ball') & (~snapshot_away['mapped id'].isin(goalkeepers))]) > len(snapshot_home[(snapshot_home['group name'] != 'Ball') & (~snapshot_home['mapped id'].isin(goalkeepers))])
        
        ## Update one shot with the calculated features
        ## the y-axis is mirrored in the shot and Kinexon Dataset => thats why i use -shot['shot_position_y']
        if not goalkeeper.empty:
            shot_list_game.at[idx, 'shooter_goalkeeper_dist'] = math.dist(np.array(snapshot_shooter[['x in m', 'y in m']]).flatten(), np.array(snapshot_goalkeeper[['x in m', 'y in m']]).flatten())
            shot_list_game.at[idx, 'shot_goalkeeper_dist'] = math.dist(np.array([shot['shot_position_x'], -shot['shot_position_y']]), np.array(snapshot_goalkeeper[['x in m', 'y in m']]).flatten())
            shot_list_game.at[idx, 'shooter_goalkeeper_angle'] = func_angle_shooter_gk(snapshot_shooter['x in m'], snapshot_shooter['y in m'], snapshot_goalkeeper['x in m'], snapshot_goalkeeper['y in m'])
            shot_list_game.at[idx, 'shot_goalkeeper_angle'] = func_angle_shooter_gk(shot['shot_position_x'], -shot['shot_position_y'], snapshot_goalkeeper['x in m'], snapshot_goalkeeper['y in m'])
            shot_list_game.at[idx, 'empty_net'] = False
        else:
            shot_list_game.at[idx, 'shooter_goalkeeper_dist'] = 0
            shot_list_game.at[idx, 'shot_goalkeeper_dist'] = 0
            shot_list_game.at[idx, 'shooter_goalkeeper_angle'] = 0
            shot_list_game.at[idx, 'shot_goalkeeper_angle'] = 0 
            shot_list_game.at[idx, 'empty_net'] = True
            
        shot_list_game.at[idx, 'player_speed'] = shooter['speed in m/s'].mean()
        
        ## Get all passes from the last shot to the actual shot
        passes_before_shot = passes[passes['ts in ms'].isin(range(last_shot_timestamp, snapshot_shot_timestamp))]
        
        idx_list_errors = []
        try:
            ## Get last ball exchange beween the teams => ball posession changes
            last_ball_exchange = passes_before_shot.loc[passes_before_shot['group id'].shift() != passes_before_shot['group id']].iloc[-1]
            ## Get all passes from last ball exchange to the shot
            passes_before_shot = passes_before_shot[passes_before_shot['ts in ms'].isin(range(last_ball_exchange['ts in ms'], shot['merge_time_shots']))]
            ## Reduce the number of records over the merge_time_shot => idea that only every second the player with posession of the ball can change  => can be really inaccurate
            ## Note: the Kinexon data contains multiple players with the same timestamp and possession of the ball
            passes_before_shot = passes_before_shot.loc[passes_before_shot['merge_time_shots'].shift() != passes_before_shot['merge_time_shots']]
            shot_list_game.at[idx, 'number_of_passes'] = len(passes_before_shot.loc[passes_before_shot['full name'].shift() != passes_before_shot['full name']])
            ## Calculate ball posession for the team in seconds
            shot_list_game.at[idx, 'ball_posession'] = (shot['merge_time_shots'] - last_ball_exchange['ts in ms'])/1e3
        except Exception:
            idx_list_errors.append(idx)
            print('Error')
                                           
        last_shot = snapshot_shot_timestamp
        # break ## comment this line in to check just the 1 shot (first shot)
    
    shot_list_df = pd.concat([shot_list_df, shot_list_game])
# display(shot_list_df)
shot_list_df.to_csv('/home/sc.uni-leipzig.de/az091jwou/expected_goals_handball/Data/detailed_shots_dataset_new.csv')
```

```python
## Import game list and shot list
path_games = '/work/users/ca439qyki/R/Kinexon/Data/Goals/'
simple_game_list_df = pd.read_csv('{path}{filename}'.format(path=path_games, filename='game_list.csv'))
simple_shot_list_df = pd.read_csv('{path}{filename}'.format(path=path_games, filename='shot_list.csv'))

## only get matches with available positional data
simple_shot_list_df = simple_shot_list_df[simple_shot_list_df['match_id'].isin(simple_game_list_df['match_id'])]

## drop unused columns and not filled rows
simple_game_list_df = simple_game_list_df.drop(simple_game_list_df.columns[0], axis=1)
simple_shot_list_df = simple_shot_list_df.dropna(subset=['shot_position_y', 'shot_position_x','game_clock']) ## drop all rows where no shot coordinates were given and pre game shots
simple_shot_list_df = simple_shot_list_df[((simple_shot_list_df['game_clock']!='30:00') &(simple_shot_list_df['validated'] != 0))] ## drop all shots from the halftime

## add columns to connect shots to kinexon data before splitting it up per game
simple_shot_list_df['timestamp_dt'] = pd.to_datetime(simple_shot_list_df['timestamp'])
simple_shot_list_df['merge_time_shots'] = simple_shot_list_df['timestamp_dt'].apply(lambda x: int(x.value / 1000000))

## add shots per game to the shot row in the dataframe
simple_game_list_df['shots'] = simple_game_list_df['match_id'].apply(lambda x: simple_shot_list_df[simple_shot_list_df['match_id'] == x])

simple_shot_list_df = pd.DataFrame()
for idx, row in simple_game_list_df.iterrows():
    ## Create features for each shot
    shot_list_game = pd.DataFrame(row['shots'])
    if shot_list_game.empty:
        continue
    
    # shot_list_game = shot_list_game[shot_list_game['merge_time_shots'] >= starting_times.get(row['match_id'])]
    shot_list_game['halftime'] = pd.to_timedelta(['00:'+time for time in shot_list_game['game_clock']]) > datetime.timedelta(minutes=30)
    shot_list_game['home'] = shot_list_game['team'].isin([row['home_team']])
    shot_list_game['norm_x_shot'], shot_list_game['norm_y_shot'] = func_coord(shot_list_game['home'], shot_list_game['halftime'], shot_list_game['shot_position_x'], shot_list_game['shot_position_y'], row['saison'])
    shot_list_game['visible_angle'] = func_visible_angle(shot_list_game['norm_x_shot'], shot_list_game['norm_y_shot'])
    shot_list_game['angle_to_mid'] = func_angle_to_mid(shot_list_game['norm_x_shot'], shot_list_game['norm_y_shot'])
    shot_list_game['score_home'], shot_list_game['score_away'] = func_game_score(shot_list_game['home'], shot_list_game['success'])
    shot_list_game['score_home'] = shot_list_game.groupby(['score_home']).ngroup().cumsum()
    shot_list_game['score_away'] = shot_list_game.groupby(['score_away']).ngroup().cumsum()
    shot_list_game.sort_values(['timestamp_dt', 'home'], inplace=True)
    shot_list_game['time_between_shots_s'] = shot_list_game.groupby('home')['timestamp_dt'].diff().dt.total_seconds()
    shot_list_game.sort_index(inplace=True)
    # display(shot_list_game)
    simple_shot_list_df = pd.concat([simple_shot_list_df, shot_list_game])
    
simple_shot_list_df.to_csv('/home/sc.uni-leipzig.de/az091jwou/expected Goals handball/Data/simple_shots_dataset.csv')
```
