```python
display(game_list_df)

kinexon_inertial_path = '/work/users/ca439qyki/R/Kinexon/Data/RAW/{short}/Saison_{season}/Spieltag/{gameday_number:02d}/{team_name_home}_vs._{team_name_away}_Match_inertial.csv'
kinexon_positions_path = '/work/users/ca439qyki/R/Kinexon/Data/RAW/{short}/Saison_{season}/Spieltag/{gameday_number:02d}/{team_name_home}_vs._{team_name_away}_Match_positions.csv'
```

```python
def round_to_next_n(x, n=50):
    return int(n * round(float(x)/n))
```

```python
players_sensors = []
for idx, row in game_list_df.iterrows():
    if row['saison'] == '2019_2020':
        teams_dict = {'SC DHFK Leipzig':('DHFK', 'SC_DHfK_Leipzig'),'Bergischer HC': ('BHC', 'Bergischer')}
    else:
        teams_dict = {'SC DHFK Leipzig':('DHFK', 'SC_DHfK_Leip'),'Bergischer HC': ('BHC', 'Bergischer_HC')}
        
    home_team = teams_dict.get(row['home_team'])
    away_team = teams_dict.get(row['away_team'])
    home_inertial_df = pd.read_csv(kinexon_inertial_path.format(short=home_team[0], season=row['saison'],gameday_number=row['game_day'], team_name_home=home_team[1], team_name_away=away_team[1]), sep=';')
    home_pos_df = pd.read_csv(kinexon_positions_path.format(short=home_team[0], season=row['saison'],gameday_number=row['game_day'], team_name_home=home_team[1], team_name_away=away_team[1]), sep=';')
    away_inertial_df = pd.read_csv(kinexon_inertial_path.format(short=away_team[0], season=row['saison'],gameday_number=row['game_day'], team_name_home=home_team[1], team_name_away=away_team[1]), sep=';')
    away_pos_df = pd.read_csv(kinexon_positions_path.format(short=away_team[0], season=row['saison'],gameday_number=row['game_day'], team_name_home=home_team[1], team_name_away=away_team[1]), sep=';')
    
    # home_inertial_df = home_inertial_df.drop(home_inertial_df.columns[[1,3,4,5,6,7,8,12,13,14,15,16,17,18]], axis=1)
    # home_pos_df = home_pos_df.drop(home_pos_df.columns[[6,12,13,14,15,18]], axis=1)       
    # away_inertial_df = away_inertial_df.drop(away_inertial_df.columns[[1,3,4,5,6,7,8,12,13,14,15,16,17,18]], axis=1)
    # away_pos_df = away_pos_df.drop(away_pos_df.columns[[6,12,13,14,15,18]], axis=1)

    away_inertial_df['merge_time'] = away_inertial_df['ts in ms'].apply(lambda x: round_to_next_n(x, n=50))
    away_pos_df['merge_time'] = away_pos_df['ts in ms']
    home_inertial_df['merge_time'] = home_inertial_df['ts in ms'].apply(lambda x: round_to_next_n(x, n=50))
    home_pos_df['merge_time'] = home_pos_df['ts in ms']
    away_game_data = pd.merge(away_pos_df, away_inertial_df, how='left', on=['merge_time', 'sensor id'], suffixes=(None, '_inertial'))
    home_game_data = pd.merge(home_pos_df, home_inertial_df, how='left', on=['merge_time', 'sensor id'], suffixes=(None, '_inertial'))
    game_data = pd.concat([away_game_data, home_game_data], ignore_index=True)

    players_sensors += zip(game_data['full name'].unique(), game_data['mapped id'].unique(), [row['saison'] for i in range(len(game_data['sensor id'].unique()))], game_data['sensor id'].unique())
```

```python
players_sensors_df = pd.DataFrame(players_sensors)
players_sensors_df = players_sensors_df.rename({0: 'Name', 1: 'mapped_id', 2:'saison', 3: 'sensor_id'}, axis=1) 

names = list(players_sensors_df['Name'].unique())
names.remove('Ball1 Ball')
names.remove('Ball2 Ball')
names.remove('Ball3 Ball')
players_sensors_df = players_sensors_df[players_sensors_df['Name'].isin(names)]
```

```python
players_sensors_map2_df = pd.merge(players_sensors_map_df, players_sensors_df, how='left', left_on='name', right_on='Name')
display(players_sensors_map2_df)
```
