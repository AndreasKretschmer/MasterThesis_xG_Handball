# Define Functions for Visualization  

```python
def display_means_per_groupdataframe(dataframe, group):
    display(dataframe.groupby(group).mean(numeric_only=True))

def plot_countplot(y, data):
    sns.countplot(x=y, data=data, palette='hls')
    plt.show()

def plot_correlation_matrix(dataframe):
    corr = dataframe.corr(method='pearson', numeric_only=True)
    sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
    plt.show()
    
def scatterplot_all_feature(dataframe):
    for feature in dataframe.columns:
        if feature in ('success','player_id','game_clock','shot_type','norm_y_shot', 'norm_x_shot'):
            continue
        sns.scatterplot(data=dataframe, x=feature, y='success', palette='RdBu')
        plt.show()
    
def plot_percent_goal_vs_feature(dataframe,feature, type_chart = "bar"):
    dataframe_copy = dataframe.copy() 
    feature_x = feature
    feature_y = 'success'
    feature_agg = 'Unnamed: 0.1'
    attrs = [feature_x, feature_y, feature_agg]
    
    if (dataframe_copy[feature_x].dtype == np.float64) and (feature_x in ('distance','speed_ball')):
        dataframe_copy[feature_x] = dataframe_copy[feature_x].apply(lambda x: round(x * 2) / 2)
    else:
        if (dataframe_copy[feature_x].dtype == np.float64) and (feature_x in ('visible_angle','angle_to_mid')):
            dataframe_copy[feature_x] = dataframe_copy[feature_x].round(1)
        else:
            if (dataframe_copy[feature_x].dtype == np.float64) and (feature_x in ('time_between_shots_s')):
                dataframe_copy[feature_x] = dataframe_copy[feature_x].round()

    outcomes_by_feature = dataframe_copy.reset_index()[attrs] \
                            .groupby([feature_x, feature_y])  \
                            .count()                          \
                            .unstack()                        \
                            .fillna(0)
    
    outcomes_by_feature = outcomes_by_feature[feature_agg].reset_index()
    outcomes_by_feature["Percent Goal"] = outcomes_by_feature[1] * 100  / outcomes_by_feature.sum(axis = 1)
    outcomes_by_feature = outcomes_by_feature.rename(columns={0: 'No Goal', 1: 'Goal'})
    
    if type_chart == "line":
        chart = alt.Chart(outcomes_by_feature).mark_line()
    elif type_chart == "bar":
        chart = alt.Chart(outcomes_by_feature).mark_bar()
    
    chart = chart.encode(
        alt.X('{}:N'.format(feature_x), axis = alt.Axis(title = feature_x, labelAngle = 0)),
        alt.Y('Percent Goal:Q', axis = alt.Axis(title = 'Percent Goal'))).properties(width = 850, height = 300, title = "Percent Goal vs {}".format(feature_x)
    ).configure_axis(labelFontSize = 13, titleFontSize = 15
    ).configure_title(fontSize = 16)
    
    display(chart)
    
def visualize_dataset(dataframe):
    y = dataset['success']
    x = dataset.drop('success', axis=1)
    plot_countplot(y, x)
    # scatterplot_all_feature(dataframe)
    plot_percent_goal_vs_feature(dataframe, 'distance')
    
    
```
